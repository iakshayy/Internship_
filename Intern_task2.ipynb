# Filtering Data

# Filtering Data where "Y-Kappa" is less than 18

```python
dataframe4 = dataframe.loc[dataframe["Y-Kappa"]<18]
print(dataframe4)
```

#  Filtering Data where "Y-Kappa" is less than 20 and "Chiprate" is less than 20 and  "BF-CMratio" is less than 80

```python
dataframe5 = dataframe.loc[(dataframe["Y-Kappa"]<20 ) & (dataframe["ChipRate"]<20) & (dataframe["BF-CMratio"]<80)]
print(dataframe5)
```

# Handling Null Values

```python
dataframe6 = dataframe.isnull().sum()
print(dataframe6)
```

# Total Number Of Null Values

```python
dataframe7 = dataframe.isnull().sum().sum()
print(dataframe7)
```

# Removing Null Values

```python
dataframe8 = dataframe.dropna()
print(dataframe8)
```

# Filling Null Values With 0

```python
dataframe9 = dataframe.fillna(value = 0)
print(dataframe9)
```

# Dropping Duplicate values that are null

```python
dataframe0= dataframe.drop_duplicates()
print(dataframe0)
```
